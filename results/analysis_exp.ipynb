## Helper function

```python
import datetime
def clean_datatime(value):
        if isinstance(value, datetime.datetime):          
            return value.month + value.day * 0.01
        elif isinstance(value, str):
            return float(value.replace('٫', '.'))
            #return float(value)
        else:
            return value
        
def clean_columns_datatime(data):
    for column in numeric_target_columns:#['Seeds', 'Total Reward', 'Waiting', 'Speed','Depart Delay', 'Time Loss', 'Waiting Car', 'Total Time(m)']:
        if column in data.columns:
            data.loc[:, column] = data[column].apply(lambda x: clean_datatime(x))
    return data

def get_grouped_seed(data): ## seeds are after each other on same experiments
    df = data.reset_index(drop=True)
    result_rows = []
    i = 0

    while i < len(df) - 2:
        group = df.iloc[i:i+3]
        if set(group['Seed']) == {0, 1, 2}:  # Check if Seeds are 0, 1, 2
            # Average numeric columns
            numeric_cols = group.select_dtypes(include='number').columns
            avg_numeric = group[numeric_cols].mean()
            
            # Keep non-numeric columns from the first row
            non_numeric_cols = group.select_dtypes(exclude='number').columns
            non_numeric = group.iloc[0][non_numeric_cols]
            
            # Combine into a single averaged row
            new_row = pd.concat([non_numeric, avg_numeric])
            result_rows.append(new_row)
            
            i += 3  # Skip the group of 3 rows
        else:
            result_rows.append(df.iloc[i])  # Add this row as is
            i += 1

    # Add any remaining rows that were not part of a group
    while i < len(df):
        result_rows.append(df.iloc[i])
        i += 1

    # Create the final DataFrame
    result_df = pd.DataFrame(result_rows)

    # Now result_df contains the modified DataFrame with averaged rows
    return result_df


```

```python
def prepare_data_benechmark(data):
    '''
    returns data,data_groubed_Seed
    '''
    #clean numeric datatime
    data = clean_columns_datatime(data)

    # makedata groubed on seed
    data_groubed_Seed = get_grouped_seed(data)
    return data,data_groubed_Seed
```

```python
import matplotlib.pyplot as plt
def plot_data_around_seed(data,data_groubed,column_x,column_y, title='',label='',marker='--',color='blue'):
    
    for i in range(3):
        plt.plot(data[data['Seed']==i][column_x],data[data['Seed']==i][column_y], label =f'seed {i} with {label}',linestyle=marker,alpha=.7,linewidth=1.5)
        #print(f"Plotting seed {i} for {column_y}, length={len(data[data['Seeds']==i])}")

    plt.plot(data_groubed[column_x],data_groubed[column_y],label ='average'+label,linestyle='-',color=color,linewidth=2.5)
    plt.title(title)
    plt.xlabel(column_x)
    plt.ylabel(column_y)
```

```python
import matplotlib.pyplot as plt

# Unique combinations
areas = data_exp_episode_effect['Area'].unique()
algorithms = data_exp_episode_effect['Algorithm'].unique()
combinations = [(a, alg) for a in areas for alg in algorithms]

# Layout control
n_cols = 2  # or 3 if you prefer
n_rows = len(combinations) * len(numeric_target_columns) // n_cols + 1

fig, axes = plt.subplots(n_rows, n_cols, figsize=(7 * n_cols, 4.5 * n_rows), squeeze=False)

# Flatten axes array for easier indexing
axes = axes.flatten()

plot_idx = 0
for area, algorithm in combinations:
    # Filter data
    filtered = data_exp_episode_effect[
        (data_exp_episode_effect['Area'] == area) &
        (data_exp_episode_effect['Algorithm'] == algorithm)
    ]

    if filtered.empty:
        continue

    data, data_groubed = prepare_data_benechmark(filtered)

    for column_y in numeric_target_columns:
        if plot_idx >= len(axes):
            break  # safety

        ax = axes[plot_idx]

        # Plot each seed line
        for seed in [0, 1, 2]:
            subset = data[data['Seed'] == seed]
            if column_x in subset.columns and column_y in subset.columns:
                ax.plot(subset[column_x], subset[column_y], label=f'Seed {seed}', linestyle='--', alpha=0.7)

        # Plot average line
        if column_x in data_groubed.columns and column_y in data_groubed.columns:
            ax.plot(data_groubed[column_x], data_groubed[column_y], label='Avg', linestyle='-', color='blue', linewidth=2.5)

        # Title and labels with adjusted font sizes
        ax.set_title(f'{area} | {algorithm} | {column_y}', fontsize=12)
        ax.set_xlabel(column_x, fontsize=10)
        ax.set_ylabel(column_y, fontsize=10)
        ax.legend(fontsize=9)

        plot_idx += 1

# Hide any unused subplots
for i in range(plot_idx, len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

## Anova

```python
import pandas as pd
from scipy.stats import ttest_ind, f_oneway

def simple_stat_test(df, metric, group_col):
    groups = df[group_col].unique() ##note should each group have data length > 1
    data = [df[df[group_col] == g][metric] for g in groups]
    
    if len(groups) == 2:
        stat, p = ttest_ind(*data)
        print(f"T-test between {groups[0]} and {groups[1]}: p = {p:.4f}")
    elif len(groups) > 2:
        stat, p = f_oneway(*data)
        print(f"ANOVA across groups in '{group_col}': p = {p:.4f}")
    else:
        print("Not enough groups to test.")

# Example usage:
# simple_stat_test(df, "Reward", "Algorithm")
```

```python
# Sample dummy data
df = pd.DataFrame({
    'Algorithm': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Reward': [100, 110, 90, 85, 105, 108]
})

simple_stat_test(df, "Reward", "Algorithm") #A p-value < 0.05 is typically considered statistically significant.


```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
#  test is a post-hoc test used after an ANOVA to perform pairwise comparisons between the groups to determine which specific groups are significantly different 

# Apply Tukey's HSD test
tukey = pairwise_tukeyhsd(endog=df["Reward"], groups=df["Algorithm"], alpha=0.05)

# Print the results
print(tukey)
```

```python
import pandas as pd
from scipy.stats import f_oneway, ttest_ind
from statsmodels.stats.multicomp import pairwise_tukeyhsd

def run_statistical_tests(df, metric, group_col):
    unique_groups = df[group_col].unique()

    print(f"\n--- Statistical Test for: {metric} grouped by {group_col} ---")

    if len(unique_groups) == 2:
        # Two groups → use t-test
        group1, group2 = unique_groups
        data1 = df[df[group_col] == group1][metric]
        data2 = df[df[group_col] == group2][metric]
        stat, pval = ttest_ind(data1, data2, equal_var=False)
        print(f"t-test between {group1} and {group2}: p-value = {pval:.4f}")
    elif len(unique_groups) > 2:
        # More than two groups → use ANOVA
        groups = [df[df[group_col] == g][metric] for g in unique_groups]
        stat, pval = f_oneway(*groups)
        print(f"ANOVA across {group_col} groups: p-value = {pval:.4f}")

        if pval < 0.05:
            print("Significant difference detected — running Tukey HSD test:")
            tukey = pairwise_tukeyhsd(endog=df[metric], groups=df[group_col], alpha=0.05)
            print(tukey)
        else:
            print("No significant differences found.")
    else:
        print("Not enough groups for comparison.")

# 🔧 Usage example:
# Replace `your_df` with your DataFrame
# Run on metric "Reward", grouped by "Algorithm"

run_statistical_tests(data_exp_episode_effect, metric="Reward", group_col="Algorithm")
```
